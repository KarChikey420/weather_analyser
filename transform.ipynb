# Convert Timestamp and Extract Components

```python
df['datetime'] = pd.to_datetime(df['datetime'])
df['date'] = df['datetime'].dt.date
df['time'] = df['datetime'].dt.time
df['hour'] = df['datetime'].dt.hour
df['minute'] = df['datetime'].dt.minute
df['weekday'] = df['datetime'].dt.day_name()
df['month'] = df['datetime'].dt.month_name()
```

```python
df.head()
```

# Kelvin to Celsius 

```python
df['temperature_celsius'] = df['temperature'] - 273.15
df['temperature_celsius'] = df['temperature_celsius'].round(2)
df.head()
```

# is_hot_day → if temperature > 30°C

# is_rainy → if weather contains “Rain”

# is_night → if hour between 18–6


```python
df['is_hot_day'] = df['temperature_celsius'] > 30
df['is_rainy'] = df['weather_category'].str.contains("Rain", case=False)
df['is_night'] = df['hour'].apply(lambda x: x < 6 or x >= 18)
df.head()
```
